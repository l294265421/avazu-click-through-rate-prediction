```python
import csv
import collections
def line_count(filename):
    result = 0
    for i, row in enumerate(csv.DictReader(open(filename)), start=1):
       result += 1
    return result
```

```python
import numpy as np
def column_feature_and_click_num(column):
    df = pd.read_csv(train_file_path, usecols=[column, 'click'])
    agg_df = df.groupby(column, as_index=False).agg({'click': [np.sum, np.size]})
    agg_df.columns = [column, 'sum', 'size']
    agg_df['ratio'] = agg_df['sum'] / agg_df['size']
    return agg_df.sort_values(by='ratio', axis=0, ascending=False)
```
