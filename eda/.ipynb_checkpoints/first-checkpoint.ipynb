```python
import csv
import collections
def line_count(filename):
    result = 0
    for i, row in enumerate(csv.DictReader(open(filename)), start=1):
       result += 1
    return result
```

```python
print('训练集size:{}'.format(line_count(train_file_path)))
print('测试集size:{}'.format(line_count(test_file_path)))
```

```python
def feature_num(columns, file_path):
    for column in columns:
        print(column + ' feature num:{}'.format(len(pd.read_csv(file_path, usecols=[column])[column].unique())))
```

```python
import numpy as np
def column_feature_and_click_num(column):
    df = pd.read_csv(train_file_path, usecols=[column, 'click'])
    agg_df = df.groupby(column, as_index=False).agg({'click': [np.sum, np.size]})
    agg_df.columns = [column, 'sum', 'size']
    agg_df['ratio'] = agg_df['sum'] / agg_df['size']
    return agg_df.sort_values(by='ratio', axis=0, ascending=False)
```

```python
column_feature_and_click_num('banner_pos')
```
